```python
def score_verb_sim(x):
    container = []
    for v in set_verbs:
        if v.lower() != x.lower():
            sim_score = leven.similarity(x.lower(), v.lower())
            sim_score = round(sim_score, 3)
            container.append((v, sim_score))
    return container
```

```python
df_verbs_choices = pd.DataFrame(data={'VERB': list(set_verbs)})
```

```python
df_verbs_choices['SOCRES'] = df_verbs_choices['VERB'].apply(lambda x: score_verb_sim(x))
```

```python
df_verbs_choices
```

```python
df_verbs_choices['SCORE_1'] = df_verbs_choices['SOCRES'].apply(lambda x: sorted(x, key=lambda tup: tup[1], reverse=True))
```

```python
def filter_choices(v, x):
    choices = []
    for i in x:
        if (i[0].lower() not in choices) and (dict_lemma_text[i[0].lower()] != dict_lemma_text[v.lower()]):
            choices.append(i[0].lower())
    return choices
```

```python
df_verbs_choices['CHOICES'] =  df_verbs_choices.apply(lambda x: filter_choices(x['VERB'], x['SCORE_1']), axis=1)
```

```python
df_noun_choices = pd.DataFrame(data={'NOUN': list(set_nouns)})
df_noun_choices['SCORES'] = df_noun_choices['NOUN'].apply(lambda x: score_verb_sim(x))
df_noun_choices['SCORE_1'] = df_noun_choices['SCORES'].apply(lambda x: sorted(x, key=lambda tup: tup[1], reverse=True))
```

```python
def make_unique_choices(words_choices):
    container = []
    lemma_check = []
    while len(container) <= 10:
        for c in words_choices:
            if dict_lemma_text[c.lower()] not in lemma_check:
                container.append(c.lower())
                lemma_check.append(dict_lemma_text[c.lower()])
    return container
```
